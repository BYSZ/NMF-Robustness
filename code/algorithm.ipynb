```python
import numpy as np
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from collections import Counter
from scipy import misc
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.metrics import normalized_mutual_info_score
#####################################################################################
#Loading data
def load_data(root='data/CroppedYaleB', reduce=4):

    images, labels = [], []

    for i, person in enumerate(sorted(os.listdir(root))):
        
        if not os.path.isdir(os.path.join(root, person)):
            continue
        
        for fname in os.listdir(os.path.join(root, person)):    
            
            # Remove background images in Extended YaleB dataset.
            if fname.endswith('Ambient.pgm'):
                continue
            
            if not fname.endswith('.pgm'):
                continue
                
            # load image.
            img = Image.open(os.path.join(root, person, fname))
            img = img.convert('L') # grey image.

            # reduce computation complexity.
            img = img.resize([s//reduce for s in img.size])

            # convert image to numpy array.
            img = np.asarray(img).reshape((-1,1))
            
            # Preprocessing
            # Normalize to the range 0-1
            img = img  / 255.0

            # collect data and label.
            images.append(img)
            labels.append(i)

    # concate all images and labels.
    images = np.concatenate(images, axis=1)
    labels = np.array(labels)

    return images, labels
#####################################################################################
#Reshape image
def restore_2d(array, w=128, h=128):

    return np.asarray(array).reshape(w, h)
#####################################################################################
#Generate Gaussian noise
def gaussian_noise(image, mean=0, sigma=0.5):
  
  return np.clip(image + np.random.normal(mean,sigma,(image.shape)), 0, 1)
#####################################################################################
#Assign cluster labels
def assign_cluster_label(X, Y):
  kmeans = KMeans(n_clusters=len(set(Y))).fit(X)
  Y_pred = np.zeros(Y.shape)
  for i in set(kmeans.labels_):
      ind = kmeans.labels_ == i
      Y_pred[ind] = Counter(Y[ind]).most_common(1)[0][0] 
  return Y_pred
#####################################################################################
#Define evaluation function
def accuracy(Y_true, Y_pred):

  return accuracy_score(Y_true, Y_pred)

def nmi_score(Y_true, Y_pred):

  return normalized_mutual_info_score(Y_true, Y_pred)

def evaluate(X, X_rec, Y_true, H):
  Y_pred = assign_cluster_label(H.T, Y_true)
  acc_metric = accuracy(Y_true, Y_pred)
  nmi_metric = nmi_score(Y_true, Y_pred)
  return acc_metric, nmi_metric
#####################################################################################
#L2,1-Norm Robust NMF
class L21normRobust_NMF():
  def __init__(self,):
    pass

  def _initialize_matrices(self, X):

    # Set seed
    rng = np.random.RandomState(self.seed)

    U = np.abs(np.random.normal(loc=0, scale=3, size=(self.m, self.k)))
    V = np.abs(np.random.normal(loc=0, scale=3, size=(self.k, self.n)))

    return U, V

  def _compute(self, X):

    U,V = self._initialize_matrices(X)
    tol=1e-5

    acc_metric = []
    nmi_metric = []

    for step in range(self.n_iter):

      D = np.diag(1 / np.sqrt(np.sum(np.square(X - U.dot(V)), axis=0)))

      Uu = U * (X.dot(D).dot(V.T))/(U.dot(V).dot(D).dot(V.T)+1e-10)
      Vu = V * (Uu.T.dot(X).dot(D))/(Uu.T.dot(Uu).dot(V).dot(D)+1e-10)
      
      if step % 5 == 0:
        acc_vle, nmi_vle = evaluate(self.X, self.reconstruct_data(Uu, Vu), self.Y, Vu)
        print(f"Iteration: {step} - Acc: {round(acc_vle, 3)} - NMI: {round(nmi_vle, 3)}")
        acc_metric.append((step, round(acc_vle, 3)))
        nmi_metric.append((step, round(nmi_vle, 3)))

      e_U = np.sqrt(np.sum((Uu-U)**2, axis=(0,1)))/U.size
      e_V = np.sqrt(np.sum((Vu-V)**2, axis=(0,1)))/V.size

      if e_U<tol and e_V<tol:
        print("step is:",step)
        break

      U = Uu
      V = Vu

    return U, V, acc_metric, nmi_metric

  def fit(self, X, 
          X_Noise,
          Y,
          n_components, 
          n_iter=8000,
          seed=24, 
          plot=True):

    # Initialize data
    self.X = X
    self.Y = Y
    self.m, self.n = X.shape
    self.k = n_components
    self.n_iter = n_iter
    self.seed = seed

    U, V, acc_metric, nmi_metric = self._compute(X_Noise)
    
    if plot:
      fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12,3))
      axs[0].plot([el[0] for el in acc_metric], [el[1] for el in acc_metric])
      axs[1].plot([el[0] for el in nmi_metric], [el[1] for el in nmi_metric])

      axs[0].set_title('Average Accuracy')
      axs[0].set_xlabel('Iterations')
      axs[0].set_ylabel('ACC')

      axs[1].set_title('Normalized Mutual Information')
      axs[1].set_xlabel('Iterations')
      axs[1].set_ylabel('NMI')
      fig.tight_layout()
    return U, V

  def reconstruct_data(self, U, V):

    return U.dot(V)
#####################################################################################
#L2-Norm MUR NMF
class L2normMUR_NMF():
  def __init__(self):
      pass

  def reconstruct_data(self, W, H):

      return W.dot(H)

  def SVDbased_init(self, X, rank):

      u,s,v=np.linalg.svd(X,full_matrices=False)
      v=v.T
      w=np.zeros((X.shape[0],rank))
      h=np.zeros((rank,X.shape[1]))

      w[:,0]=np.sqrt(s[0])*np.abs(u[:,0])
      h[0,:]=np.sqrt(s[0])*np.abs(v[:,0].T)

      for i in range(1,rank):
          
          ui=u[:,i]
          vi=v[:,i]
          ui_pos=(ui>=0)*ui
          ui_neg=(ui<0)*-ui
          vi_pos=(vi>=0)*vi
          vi_neg=(vi<0)*-vi
          
          ui_pos_norm=np.linalg.norm(ui_pos,2)
          ui_neg_norm=np.linalg.norm(ui_neg,2)
          vi_pos_norm=np.linalg.norm(vi_pos,2)
          vi_neg_norm=np.linalg.norm(vi_neg,2)
          
          norm_pos=ui_pos_norm*vi_pos_norm
          norm_neg=ui_neg_norm*vi_neg_norm
          
          if norm_pos>=norm_neg:
              w[:,i]=np.sqrt(s[i]*norm_pos)/ui_pos_norm*ui_pos
              h[i,:]=np.sqrt(s[i]*norm_pos)/vi_pos_norm*vi_pos.T
          else:
              w[:,i]=np.sqrt(s[i]*norm_neg)/ui_neg_norm*ui_neg
              h[i,:]=np.sqrt(s[i]*norm_neg)/vi_neg_norm*vi_neg.T
      return w, h

  def random_init(self, X, rank):

      number_of_features = X.shape[0]
      number_of_observations = X.shape[1]

      W = np.random.uniform(1,2,(number_of_features,self.rank))
      H = np.random.uniform(1,2,(self.rank,number_of_observations))
      return W, H

  # L2 Multiplicative NMF
  def compute(self, X):
         
      W, H = self.SVDbased_init(X, self.rank)
     #W, H = self.random_init(X, self.rank)
      k = self.rank
      e = 1.0e-10
      
      norms = []
      acc_metric = []
      nmi_metric = []

      for n in range(self.n_iter):
          # Update H
          W_TX = W.T@X
          W_TWH = W.T@W@H+e
          for i in range(np.size(H, 0)):
              for j in range(np.size(H, 1)):
                  H[i, j] = H[i, j] * W_TX[i, j] / W_TWH[i, j]
          # Update W
          XH_T = X@H.T
          WHH_T =  W@H@H.T+ e

          for i in range(np.size(W, 0)):
              for j in range(np.size(W, 1)):
                  W[i, j] = W[i, j] * XH_T[i, j] / WHH_T[i, j]

          norm = np.linalg.norm(X - W@H, 'fro')
          norms.append(norm)
      
          # Print loss
          if n % 5 == 0:
            # Calculate Metrics
            acc_vle, nmi_vle = evaluate(self.X, self.reconstruct_data(W, H), 
                                                self.Y, H)

            acc_metric.append((n, round(acc_vle, 3)))
            nmi_metric.append((n, round(nmi_vle, 3)))
            print(f"Iteration: {n} - Acc: {round(acc_vle, 3)} - NMI: {round(nmi_vle, 3)}")
      return W, H,  acc_metric, nmi_metric

  def fit(self, X, X_Noise, Y, n_components,
          n_iter, 
          plot=True):

      # Initialize data
      self.X = X
      self.X_Noise = X_Noise
      self.Y = Y
      self.n_iter = n_iter
      self.rank = n_components

      W, H, acc_metric, nmi_metric = self.compute(X_Noise)

      if plot:
        fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
        axs[0].plot([el[0] for el in acc_metric], [el[1] for el in acc_metric])
        axs[1].plot([el[0] for el in nmi_metric], [el[1] for el in nmi_metric])

        axs[0].set_title('Average Accuracy')
        axs[0].set_xlabel('Iterations')
        axs[0].set_ylabel('ACC')

        axs[1].set_title('Normalized Mutual Information')
        axs[1].set_xlabel('Iterations')
        axs[1].set_ylabel('NMI')
        fig.tight_layout()
      return W, H
#####################################################################################
#Run!
def process_dataset(root, reduce_factor, noise_sigma, n_components, dataset_name):
    # Reading dataset
    X, Y = load_data(root=root, reduce=reduce_factor)
    img_size = [i // reduce_factor for i in (92, 112)] if 'ORL' in dataset_name else [i // reduce_factor for i in (168, 192)]
    
    print(f"Shape X: {X.shape}, Shape Y: {Y.shape}")
    
    # Adding gaussian noise to the original dataset
    X_Noise = gaussian_noise(X, mean=0, sigma=noise_sigma)
    
    # Initialize algorithm (L21normRobust_NMF)
    L21 = L21normRobust_NMF()
    U, V = L21.fit(X=X, X_Noise=X_Noise, Y=Y, n_components=n_components, n_iter=30)
    
    # Reconstruct image
    X_hat = L21.reconstruct_data(U, V)
    
    # Evaluation metrics
    acc_metric, nmi_metric = evaluate(X=X, X_rec=X_hat, Y_true=Y, H=V)
    print(f"Average Accuracy (L21): {acc_metric}")
    print(f"Normalized mutual information (L21): {nmi_metric}")
    
    # Plot results
    ind = 5 if 'ORL' in dataset_name else 13
    plot_images(X, X_Noise, X_hat, img_size, ind)
    
    # Initialize algorithm (L2normMUR_NMF)
    L2 = L2normMUR_NMF()
    W, H = L2.fit(X=X, X_Noise=X_Noise, Y=Y, n_components=n_components, n_iter=30)
    
    # Reconstruct image
    X_hat = L2.reconstruct_data(W, H)
    
    # Evaluation metrics
    acc_metric, nmi_metric = evaluate(X=X, X_rec=X_hat, Y_true=Y, H=H)
    print(f"Average Accuracy (L2): {acc_metric}")
    print(f"Normalized mutual information (L2): {nmi_metric}")
    
    # Plot results
    plot_images(X, X_Noise, X_hat, img_size, ind)

#Image Output!
def plot_images(X, X_Noise, X_hat, img_size, ind):
    fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(10, 3))
    plt.subplots_adjust(hspace=0.8)

    org_data = restore_2d(X[:, ind], h=img_size[0], w=img_size[1])
    axs[0].imshow(org_data, cmap=plt.cm.gray)
    axs[0].set_xticks([])
    axs[0].set_yticks([])
    axs[0].set_title('Image(Original)')

    noise_data = restore_2d(X_Noise[:, ind], h=img_size[0], w=img_size[1])
    axs[1].imshow(noise_data, cmap=plt.cm.gray)
    axs[1].set_xticks([])
    axs[1].set_yticks([])
    axs[1].set_title('Image(Noise)')

    rec_data = restore_2d(X_hat[:, ind], h=img_size[0], w=img_size[1])
    axs[2].imshow(rec_data, cmap=plt.cm.gray)
    axs[2].set_xticks([])
    axs[2].set_yticks([])
    axs[2].set_title('Image(Reconstructed)')

    plt.show()
#####################################################################################
# Process ORL dataset
process_dataset(root='data/ORL', reduce_factor=3, noise_sigma=0.2, n_components=15, dataset_name='ORL')

# Process Extended YaleB dataset
process_dataset(root='data/CroppedYaleB', reduce_factor=4, noise_sigma=0.2, n_components=15, dataset_name='Extended YaleB')
```
